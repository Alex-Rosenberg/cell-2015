### Training the model

The first thing we have to do is create a giant matrix to represent 6-mers at each position. Our matrix have $4^6 \cdot d$ columns, where d is the number of positions and n rows, where n is the number of data points:<br><br>
$AAAAAA_{1,1},\ ...\ TTTTTTT_{1,1},\ AAAAAA_{1,2},\ ... TTTTTTT_{1,2},\ ...\ AAAAAA_{1,d},\ TTTTTTT_{1,d}$<br>
$AAAAAA_{2,1},\ ...\ TTTTTTT_{2,1},\ AAAAAA_{2,2},\ ... TTTTTTT_{2,2},\ ...\ AAAAAA_{2,d},\ TTTTTTT_{2,d}$<br>
...<br>
$AAAAAA_{n,1},\ ...\ TTTTTTT_{n,1},\ AAAAAA_{n,2},\ ... TTTTTTT_{n,2},\ ...\ AAAAAA_{n,d},\ TTTTTTT_{n,d}$<br><br>
For example if AAAAAA occurs at position 1 in the first plasmid, then $AAAAAA_{1,1}$ will equal 1, while $AAAAAC_{1,1}$ through $TTTTTT_{1,1}$ will equal 0.<br><br>
To score each splice site we are going to look at a 160nt window (-80nt and +80nt). We only want to look at variable sequences, so we will look at all 6-mers that overlap the randomized regions by at least 1nt. So we will pad the edges of our matrix with zeros. The fixed sequence between the random regions will be set to zero as well. We will use the scipy sparse matrix package to make this matrix.

```python
def make_padded_nmer_matrix(seqs,mer_len):
    mer_dict = dict(zip(dnatools.make_mer_list(6),range(4**6)))
    rows,cols,vals = [],[],[]
    for i in xrange(len(seqs)):
        cur_seq = A5SS_seqs[i]
        for b in range(7-mer_len,32)+range(50-mer_len,75):
            cur_mer = cur_seq[b:b+mer_len]
            cur_mer_list_pos = mer_dict[cur_mer]
            rows.append(i)
            cols.append(cur_mer_list_pos+4**mer_len*(b+80))
            vals.append(1)
        if(i%10000)==0:
            print i,
            
        # Add a zero to the bottom right of the matrix to ensure the
        # correct size
        rows.append(len(A5SS_data)-1)
        cols.append(4**6*230-1)
        vals.append(0)
    X = scipy.sparse.csr_matrix((vals,(rows,cols)))
    return X
```

```python
X = make_padded_nmer_matrix(A5SS_seqs.values,6)
```

```python
w = data['Mer_scores'][:4**6*8]
w_mat = w.reshape(4**6,size(w)/4**6)
mer6_dict = dict(zip(dnatools.make_mer_list(6),range(4**6)))
pos_map = dict(zip(range(-80,80),[0]*40+[1]*(40-9)+[-1]*6+[2,3,4,5]+[-1]*5+[6]*29+[7]*45))
def score_seq(seq,sd):
    scores = np.zeros(90)
    for b in range(2,32)+range(45,75):
        rel_pos = b-sd
        cur_mer = seq[b:b+6]
        cur_pos_ind = pos_map[rel_pos]
        if(cur_pos_ind>=0):
            scores[b+4:b+6+4] += w_mat[mer6_dict[cur_mer],cur_pos_ind]
    return scores
```
